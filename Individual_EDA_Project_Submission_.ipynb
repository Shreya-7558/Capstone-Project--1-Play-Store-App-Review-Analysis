## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the current dataset and assigning to df and df1
df=Play_Store_Data.copy()
df1=User_Reviews.copy()
````

````python
# Filling Missing Valuesin Play Store Dataset
df['Rating'].fillna(df['Rating'].mean(),inplace = True)
df['Rating'].isnull().sum()
````

````python
# Drop Duplicates in Play Store Dataset
df.drop_duplicates(inplace = True)
df.drop_duplicates(subset='App',inplace = True)
df.duplicated().value_counts()
````

````python
# Drop all null values in Play Store Dataset
df.dropna(inplace = True)
````

````python
# Fill missing values in Sentiment in User Reviews Dataset
df1 = df1[~df1['Sentiment'].isnull()]
df1.dropna(inplace = True)
df1.isnull().sum()
````

````python
# 1. Identify the top categories with the highest number of apps
top_categories = df['Category'].value_counts().head(5)
print("Top Categories:")
print(top_categories)
````

````python
# 2. Calculate the sentiment distribution in user reviews
sentiment_distribution = df1['Sentiment'].value_counts()
print("Sentiment Distribution:")
print(sentiment_distribution)
````

````python
# 3. Analyze the distribution of app ratings
app_ratings = df['Rating']
average_rating = df.mean()
print("Average App Rating:", average_rating)
````

````python
# 4. Calculate the average sentiment polarity for each app
average_polarity = df1.groupby('App')['Sentiment_Polarity'].mean()
print("Average Sentiment Polarity by App:")
print(average_polarity)
````

````python
# 5. Analyze the relationship between app ratings and sentiment subjectivity
ratings_subjectivity = pd.merge(df[['App', 'Rating']], df1[['App', 'Sentiment_Subjectivity']], on='App')
correlation = ratings_subjectivity['Rating'].corr(ratings_subjectivity['Sentiment_Subjectivity'])
print("Correlation between App Ratings and Sentiment Subjectivity:", correlation)
````

````python
# 6. Identify apps with the most positive sentiment polarity
top_positive_apps = average_polarity.nlargest(5)
print("Top Apps with Positive Sentiment Polarity:")
print(top_positive_apps)
````
